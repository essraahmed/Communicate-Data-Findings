## Univariate Exploration


```python
#cancellation reasons
labels = 'Weather', 'Carrier','National Air System', 'Security'
plt.subplots(figsize=(14,6))
sorted_counts = flight['CancellationCode'].value_counts()
plt.title('Cancellation Reasons', fontsize=(20))
explode = (0.1, 0.1, 0.1,0.1)
plt.pie(sorted_counts ,labels= labels,autopct='%1.1f%%',startangle=90, shadow = True, explode=explode)
plt.show()
```

<ul>
<li>The most reasons of cancellation are Weather and Carrier.</li></ul>

```python
#cancelled flights
labels = 'Completed', 'Cancelled'
plt.subplots(figsize=(14,6))
sorted_counts = flight['Cancelled'].value_counts()
plt.title('Cancelled Flights', fontsize=(20))
explode = (0.1, 0.1)
plt.pie(sorted_counts ,labels= labels,autopct='%1.1f%%',startangle=90, shadow = True, explode=explode)
plt.show()
```

<ul><li>It appears that a lot of flights not cancelled</li></ul>

```python
#Arrival Delay
plt.subplots(figsize=(14,6))
plt.title('Arrival Delay', fontsize=(20))
bins = np.arange(0, flight['ArrDelay'].max()+0.5, 30)
plt.hist(data= flight, x= 'ArrDelay', bins=bins);
plt.xlim(-30,400);
```

<ul><li>The Arrival distribution is right skewed, there are flights arrived before their time.
</li></ul>

```python
#departure delay
plt.subplots(figsize=(14,6))
plt.title('Departure Delay', fontsize=(20))
bins = np.arange(0, flight['DepDelay'].max()+0.5, 30)
plt.hist(data= flight, x= 'DepDelay', bins=bins);
plt.xlim(-30,350);
```

<ul><li>The departure time delay distribution  is right skewed
</li></ul>

```python
#Number of flights
plt.subplots(figsize=(14,6))
plt.title('Number of Flights', fontsize=(20))
count_order = flight.UniqueCarrier.value_counts().index
base_color = sb.color_palette()[0]
sb.countplot(data = flight, x = 'UniqueCarrier', color = base_color, order = count_order);
plt.ylabel('Number of Flights', fontsize=(12))
plt.xlabel('UniqueCarrier', fontsize=(12))
```

<ul><li>The WN Carrier has the most number of flights.
</li></ul>

## Bivariate Exploration

> In this section, investigate relationships between pairs of variables in your
data. Make sure the variables that you cover here have been introduced in some
fashion in the previous section (univariate exploration).

```python
#relation between Arrival and Departure 
flight.plot(x = 'ArrDelay', y = 'DepDelay', kind = 'scatter', figsize=(14,6))
plt.title('Arrival vs Departure Delay', fontsize =(20))
plt.xlabel('Arrival', fontsize=(12))
plt.ylabel('Departure', fontsize=(12))
```

<ul><li>Positive correlation between Arrival and Departure Delay
</li></ul>

```python
#Arrival delay vs distance 
plt.figure(figsize=[14,6])
plt.scatter(data=flight, x="Distance", y="ArrDelay", s=1)
plt.title('Arrival Delay vs Distance', fontsize =(20))
plt.xlabel('Distance', fontsize=(12))
plt.ylabel('Arrival Delay', fontsize=(12))
plt.ylim([-100,4000]);
```

<ul><li>When the distance increases, the arrival delay decreases. 
</li></ul>

```python
#correlation between delays
variables = ['ArrDelay','DepDelay','CarrierDelay','WeatherDelay', 'NASDelay', 'SecurityDelay', 'LateAircraftDelay']
plt.figure(figsize=[14,6])
sb.heatmap(flight[variables].corr(), annot=True, fmt=".2f",  cmap = 'vlag_r', center = 0)
```

<ul><li>Security Delay has a weak relation vs other. There is a strong relation Arrival delay and departure delay.
</ul></li>

```python
#flights per Days
plt.figure(figsize=[14,6])
plt.title('Flights per Day of Week', fontsize =(20))
bins = np.arange(0, flight['DayOfWeek'].max()+0.5, 30)
base_color = sb.color_palette()[0]
days= ['Mon', 'Tue', 'Wed', 'Thur', 'Fri', 'Sat', 'Sun']
ax= sb.countplot(data= flight, x='DayOfWeek', color=base_color)
ax.set_xticklabels(days);
```

<ul><li> Wednesday and Friday is the most days of flights. Minimum number of flights on Saturday   
</li></ul>

```python
#Air Time vs Distance
plt.figure(figsize=[14,6])
base_color = sb.color_palette()[0]
flight.plot(x = 'Distance', y = 'AirTime', kind = 'scatter', color=base_color,s=10)
plt.title('AirTime vs Distance', fontsize =(20))
plt.xlabel('Distance', fontsize=(12))
plt.ylabel('AirTime', fontsize=(12));
```

<ul><li>Positive correlation between Air Time and Distance. When Distance increase the Air Time increase.
    </li></ul>

## Multivariate Exploration

> Create plots of three or more variables to investigate your data even
further. Make sure that your investigations are justified, and follow from
your work in the previous sections.

```python
g = sb.PairGrid(data=flight, vars = ['ArrDelay', 'DepDelay','WeatherDelay', 'CarrierDelay'])
g.map_diag(plt.hist)
g.map_offdiag(plt.scatter)
```

<ul><li>Weather Delay has a significant impact on Arrival and Departure Delays.
</ul></li>

```python
g=sb.FacetGrid(data=flight, col='Month', col_wrap=5)
g.map(plt.scatter,'WeatherDelay','ArrDelay')
```

<ul><li>Positive correlation between weather delay and arrival delay. Jan, Feb and Dec have the most delyas.
</li></ul>

```python
g = sb.FacetGrid(data = flight, hue = 'UniqueCarrier', height = 5)
g.map(plt.scatter, 'ArrDelay', 'DepDelay')
g.add_legend()
plt.title('UniqueCarrier with Arrival and Departure Delays', fontsize =(14))
plt.xlabel('Arrival Delay',  fontsize=(12))
plt.ylabel('Departure Delay',  fontsize=(12))
```

```python
#distance with arrival and departure delays
plt.figure(figsize=[12,6])
plt.scatter(data= flight, x='ArrDelay', y='DepDelay', c='Distance');
plt.colorbar(label='Distance')
plt.ylabel('Departure Delay', fontsize=(12))
plt.xlabel('Arrive Delay', fontsize=(12));
plt.title('Distance with Arrival and Departure Delays', fontsize =(14))
```

<ul><li>When Distance increase the arrival and departual delays increase.
</li></ul>
