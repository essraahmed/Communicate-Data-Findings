# Investigate Overview
> In this investigation on the US Flights for 2008, I will focus on: Arrival and Departual Times, Cancellation and Delay Reasons.

# Dataset Overview
>  Flight Dataset of the year 2008, contains various columns about delays and cancellation of the flights. 
Data Source: https://community.amstat.org/jointscsg-section/dataexpo/dataexpo2009 

```python
# import all packages and set plots to be embedded inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

%matplotlib inline
```

```python
#load data
flight = pd.read_csv('2008.csv')
```

## Cancellation Reasons
<ul>
<li>The most reasons of cancellation are Weather and Carrier.</li></ul>

```python
#cancellation reasons
labels = 'Weather', 'Carrier','National Air System', 'Security'
plt.subplots(figsize=(14,6))
sorted_counts = flight['CancellationCode'].value_counts()
plt.title('Cancellation Reasons', fontsize=(20))
explode = (0.1, 0.1, 0.1,0.1)
plt.pie(sorted_counts ,labels= labels,autopct='%1.1f%%',startangle=90, shadow = True, explode=explode)
plt.show()
```

<ul>
<li>The most reasons of cancellation are Weather and Carrier.</li></ul>

# Arrival Delay vs Distance
<ul><li>When the distance increases, the arrival delay decreases. 
</li></ul>

```python
#Arrival delay vs distance 
plt.figure(figsize=[14,6])
#plt.scatter(data=flight, x="Distance", y="ArrDelay", s=1, alpha=1/5)
sb.regplot(data = flight, x = 'Distance', y = 'ArrDelay', truncate=False, x_jitter=0.3, scatter_kws={'alpha':1/5});
plt.title('Arrival Delay vs Distance', fontsize =(20))
plt.xlabel('Distance(miles)', fontsize=(12))
plt.ylabel('Arrival Delay(min)', fontsize=(12))
plt.ylim([-100,3500]);
```

<ul><li>Positive correlation between Arrival and Departure Delay
</li></ul>

# Correlation between Delays
<ul><li>Security Delay has a weak relation vs other. There is a strong relation Arrival delay and departure delay.</ul></li>

```python
#correlation between delays
variables = ['ArrDelay','DepDelay','CarrierDelay','WeatherDelay', 'NASDelay', 'SecurityDelay', 'LateAircraftDelay']
plt.figure(figsize=[14,6])
plt.title('Correlation between Delays', fontsize =(20));
sb.heatmap(flight[variables].corr(), annot=True, fmt=".2f",  cmap = 'vlag_r', center = 0);
```

<ul><li>Security Delay has a weak relation vs other. There is a strong relation Arrival delay and departure delay.
</ul></li>

# Correlation between Arrival and Departure Delay
<ul><li>Positive correlation between Arrival and Departure Delay</li></ul>

```python
g = sb.FacetGrid(data = flight, hue = 'UniqueCarrier', height = 5)
g.map(plt.scatter, 'ArrDelay', 'DepDelay');
g.add_legend();
plt.title('UniqueCarrier with Arrival and Departure Delays', fontsize =(14));
plt.xlabel('Arrival Delay(min)',  fontsize=(12))
plt.ylabel('Departure Delay(min)',  fontsize=(12));
```

# Distance with Arrival and Departure Delays
<ul><li>When Distance increase the arrival and departual delays increase.
</li></ul>

```python
#distance with arrival and departure delays
plt.figure(figsize=[12,6])
plt.scatter(data= flight, x='ArrDelay', y='DepDelay', c='Distance');
plt.colorbar(label='Distance(miles)')
plt.ylabel('Departure Delay(min)', fontsize=(12))
plt.xlabel('Arrive Delay(min)', fontsize=(12));
plt.title('Distance with Arrival and Departure Delays', fontsize =(14));
```

<ul><li>When Distance increase the arrival and departual delays increase.
</li></ul>
